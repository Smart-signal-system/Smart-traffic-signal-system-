```python
"""
AI Traffic Light Controller v9.0
==================================
المنطق:
  1. الأحمر الكامل: تستنى 5 ثواني تجمع بيانات الزحمة من كل الشوارع.
  2. بعد الـ5 ثواني: تفتح الشارع الأكثر زحمة.
  3. الشارع المفتوح يفضل مفتوح لحد ما يتفرغ تماماً (0 سيارات).
  4. بعد التفريغ: أصفر 3 ثواني ← أحمر 5 ثواني (تجميع) ← الأكثر زحمة.
"""

import os, sys, random, pickle
import cv2
import numpy as np
import traci

# ─────────────────────────────────────────────────────
#  إعدادات
# ─────────────────────────────────────────────────────
SUMO_HOME   = os.environ.get("SUMO_HOME", r"C:\Program Files (x86)\Eclipse\Sumo")
sys.path.append(os.path.join(SUMO_HOME, "tools"))

PROJECT_DIR = r"C:\Users\mariam\simulation"
SUMO_CFG    = os.path.join(PROJECT_DIR, "project.sumocfg")
VIDEO_FILE  = os.path.join(PROJECT_DIR, "simulation_video.avi")
QTABLE_FILE = os.path.join(PROJECT_DIR, "qtable.pkl")

SIM_STEPS        = 3600
STEP_LEN         = 1.0
VIDEO_FPS        = 12
VIDEO_W, VIDEO_H = 1280, 720

ALPHA   = 0.4
GAMMA   = 0.90
EPSILON = 0.1

# ─── الأوقات الثابتة ───
COLLECT_TIME = 5   # ثواني تجميع بيانات الزحمة (الأحمر الكامل)
YELLOW_TIME  = 3   # ثواني الأصفر
MIN_GREEN    = 10  # الحد الأدنى للأخضر — الشارع يفضل مفتوح على الأقل 10 ثواني

# ─────────────────────────────────────────────────────
#  دوال مساعدة
# ─────────────────────────────────────────────────────
def get_total_vehicles(lane):
    try:    return traci.lane.getLastStepVehicleNumber(lane)
    except: return 0

def get_waiting_vehicles(lane):
    try:    return traci.lane.getLastStepHaltingNumber(lane)
    except: return 0

def get_waiting_time_total(lanes):
    total = 0.0
    for lane in lanes:
        try:
            for v in traci.lane.getLastStepVehicleIDs(lane):
                total += traci.vehicle.getWaitingTime(v)
        except: pass
    return total

def make_all_red(n):
    return 'r' * n


# ─────────────────────────────────────────────────────
#  كلاس العميل
# ─────────────────────────────────────────────────────
class QLAgent:
    # الأطوار
    SIG_COLLECT  = "COLLECT"    # أحمر + تجميع بيانات 5 ثواني
    SIG_GREEN    = "GREEN"      # أخضر (يفضل لحد ما يخلص)
    SIG_CLEARING = "CLEARING"   # أخضر + انتظار تفريغ كامل
    SIG_YELLOW   = "YELLOW"     # أصفر 3 ثواني

    def __init__(self, tls_id):
        self.tls_id = tls_id
        self.q_table = {}

        # نبدأ بطور التجميع مباشرة
        self.sig         = self.SIG_COLLECT
        self.phase_timer = COLLECT_TIME

        self.prev_state   = None
        self.prev_action  = 0
        self.cleared      = 0   # عداد مرات التفريغ الكامل
        self.elapsed_green = 0  # كم ثانية الشارع الحالي مفتوح
        self.log          = []

        self.controlled_lanes = list(traci.trafficlight.getControlledLanes(tls_id))
        self.num_links        = len(self.controlled_lanes)

        logic = traci.trafficlight.getAllProgramLogics(tls_id)[0]

        self.green_phases = []
        for i, ph in enumerate(logic.phases):
            has_green = ph.state.count('G') + ph.state.count('g') > 0
            has_yellow = ph.state.count('y') > 0
            if has_green and not has_yellow:
                self.green_phases.append((i, ph.state))

        self.phase_lanes = {}
        for idx, state in self.green_phases:
            self.phase_lanes[idx] = [
                self.controlled_lanes[j]
                for j, c in enumerate(state) if c in ("G", "g")
            ]

        # لا نحدد شارع هنا — هيتحدد بعد الـ5 ثواني حسب الزحمة
        self.current_phase_idx = None
        self.current_state_str = None

        # نبدأ بأحمر كامل أثناء التجميع
        traci.trafficlight.setRedYellowGreenState(tls_id, make_all_red(self.num_links))
        print(f"[{tls_id}] ✅ جاهز — بدأ تجميع بيانات الزحمة ({COLLECT_TIME} ثواني)...")

    # ─── حساب السيارات ───
    def _total(self, phase_idx):
        return sum(get_total_vehicles(l) for l in self.phase_lanes[phase_idx])

    def _waiting(self, phase_idx):
        return sum(get_waiting_vehicles(l) for l in self.phase_lanes[phase_idx])

    def _busiest_phase(self):
        """اختيار الشارع الأكثر زحمة — بيحسب كل السيارات (واقفة + متحركة)"""
        best_idx, best_str, max_w = self.green_phases[0][0], self.green_phases[0][1], -1
        for idx, state_str in self.green_phases:
            w = self._total(idx)
            if w > max_w:
                max_w, best_idx, best_str = w, idx, state_str
        return best_idx, best_str, max_w

    def _get_obs(self):
        queues = [min(self._waiting(idx), 20) for idx, _ in self.green_phases]
        return tuple(queues) + (self.current_phase_idx if self.current_phase_idx is not None else -1,)

    def _learn(self, s, a, r, s2):
        for st in (s, s2):
            if st not in self.q_table:
                self.q_table[st] = [0.0]
        old = self.q_table[s][a]
        self.q_table[s][a] = old + ALPHA * (r + GAMMA * max(self.q_table[s2]) - old)

    # ─── خطوة المحاكاة ───
    def step(self, sim_step):
        self.phase_timer -= STEP_LEN

        # ══════════════════════════════════════════════
        #  طور التجميع: أحمر كامل 5 ثواني لجمع بيانات الزحمة
        #  ⚠️ الإشارة لا تفتح أبداً قبل ما الـ5 ثواني تخلص كاملة
        # ══════════════════════════════════════════════
        if self.sig == self.SIG_COLLECT:
            if self.phase_timer > 0:
                pass  # نستنى الـ5 ثواني تخلص كاملة

            else:
                # ✅ الـ5 ثواني خلصت — دلوقتي نشوف الزحمة ونقرر
                best_idx, best_str, max_waiting = self._busiest_phase()

                self.current_phase_idx = best_idx
                self.current_state_str = best_str
                self.sig = self.SIG_GREEN
                self.phase_timer = 0
                self.elapsed_green = 0

                obs    = self._get_obs()
                action = 0
                all_lanes = [l for ls in self.phase_lanes.values() for l in ls]
                reward = -get_waiting_time_total(list(set(all_lanes)))
                if self.prev_state is not None:
                    self._learn(self.prev_state, action, reward, obs)
                self.prev_state  = obs
                self.prev_action = action

                label = "الأكثر زحمة" if max_waiting > 0 else "مافيش زحمة — افتراضي"
                entry = f"[{sim_step}s] فتح شارع {best_idx} | {label} ({max_waiting} سيارة)"
                self.log.append(entry)
                if len(self.log) > 10:
                    self.log.pop(0)

                traci.trafficlight.setRedYellowGreenState(self.tls_id, self.current_state_str)
                print(f"  [{sim_step}s] 🟢 بعد الـ{COLLECT_TIME}ث: شارع {best_idx} | {label} ({max_waiting} سيارة).")

        # ══════════════════════════════════════════════
        #  طور الأخضر
        # ══════════════════════════════════════════════
        elif self.sig == self.SIG_GREEN:
            self.elapsed_green += STEP_LEN
            total_now = self._total(self.current_phase_idx)

            if self.elapsed_green < MIN_GREEN:
                pass
            else:
                if total_now == 0:
                    self.sig = self.SIG_CLEARING
                    print(f"  [{sim_step}s] ✅ شارع {self.current_phase_idx} فضي بعد {self.elapsed_green:.0f}ث — ننتقل...")

        # ══════════════════════════════════════════════
        #  طور التأكيد
        # ══════════════════════════════════════════════
        elif self.sig == self.SIG_CLEARING:
            total_now = self._total(self.current_phase_idx)
            if total_now == 0:
                self.sig = self.SIG_YELLOW
                self.phase_timer = YELLOW_TIME
                self.cleared += 1

                yellow_state = self.current_state_str.replace('G', 'y').replace('g', 'y')
                traci.trafficlight.setRedYellowGreenState(self.tls_id, yellow_state)
                print(f"  [{sim_step}s] 🟡 تفريغ كامل #{self.cleared} — أصفر {YELLOW_TIME} ثواني.")
            else:
                self.sig = self.SIG_GREEN
                print(f"  [{sim_step}s] 🚗 رجعت سيارات ({total_now}) — رجعنا للأخضر.")

        # ══════════════════════════════════════════════
        #  طور الأصفر
        # ══════════════════════════════════════════════
        elif self.sig == self.SIG_YELLOW:
            if self.phase_timer <= 0:
                self.sig = self.SIG_COLLECT
                self.phase_timer = COLLECT_TIME

                traci.trafficlight.setRedYellowGreenState(self.tls_id, make_all_red(self.num_links))
                print(f"  [{sim_step}s] 🔴 أحمر — بنجمع بيانات الزحمة {COLLECT_TIME} ثواني...")


# ─────────────────────────────────────────────────────
#  رسم الشاشة
# ─────────────────────────────────────────────────────
def draw_frame(step, agents):
    frame = np.full((VIDEO_H, VIDEO_W, 3), (22, 22, 32), dtype=np.uint8)
    ag    = list(agents.values())[0] if agents else None

    cv2.putText(frame, "Smart Traffic v9.0 — 5s Collect + Demand-Driven",
                (30, 38), cv2.FONT_HERSHEY_SIMPLEX, 0.85, (255, 255, 255), 2)
    cv2.putText(frame, f"Step: {step}s / {SIM_STEPS}s",
                (30, 64), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (160, 200, 255), 1)

    # شريط التقدم
    prog = int((step / SIM_STEPS) * (VIDEO_W - 60))
    cv2.rectangle(frame, (30, 76), (VIDEO_W-30, 88), (55, 55, 75), -1)
    cv2.rectangle(frame, (30, 76), (30+prog, 88),    (80, 180, 80), -1)

    # التقاطع
    cx, cy, RW = VIDEO_W//2, VIDEO_H//2+20, 80
    cv2.rectangle(frame, (0, cy-RW-7),  (VIDEO_W, cy+RW+7),  (65,65,65), -1)
    cv2.rectangle(frame, (cx-RW-7, 0),  (cx+RW+7, VIDEO_H),  (65,65,65), -1)
    cv2.rectangle(frame, (0, cy-RW),    (VIDEO_W, cy+RW),     (48,48,48), -1)
    cv2.rectangle(frame, (cx-RW, 0),    (cx+RW, VIDEO_H),     (48,48,48), -1)

    # السيارات
    try:
        bbox       = traci.simulation.getNetBoundary()
        xmin, ymin = bbox[0]; xmax, ymax = bbox[1]
        sc = min((VIDEO_W-80)/max(xmax-xmin,1), (VIDEO_H-140)/max(ymax-ymin,1)) * 0.82
        ox = (VIDEO_W-(xmax-xmin)*sc)/2
        oy = (VIDEO_H-(ymax-ymin)*sc)/2+50
        for vid in traci.vehicle.getIDList()[:300]:
            vx, vy = traci.vehicle.getPosition(vid)
            spd    = traci.vehicle.getSpeed(vid)
            px = int(ox+(vx-xmin)*sc)
            py = int(VIDEO_H-(oy+(vy-ymin)*sc))
            if 4 < px < VIDEO_W-4 and 4 < py < VIDEO_H-4:
                cv2.rectangle(frame, (px-5,py-3), (px+5,py+3),
                              (50,220,50) if spd > 1.0 else (50,50,220), -1)
    except: pass

    if not ag: return frame

    # ── لوحة اليسار ──
    bx = 10
    cv2.rectangle(frame, (bx,100), (bx+335,450), (28,28,42), -1)
    cv2.rectangle(frame, (bx,100), (bx+335,450), (80,80,120), 1)
    cv2.putText(frame, "حالة الإشارة",
                (bx+8,124), cv2.FONT_HERSHEY_SIMPLEX, 0.52, (255,220,60), 1)

    if ag.sig == QLAgent.SIG_COLLECT:
        txt = f"🔴 تجميع زحمة — متبقي {max(0,ag.phase_timer):.0f}ث"
        col = (60,60,220)
    elif ag.sig == QLAgent.SIG_GREEN:
        total = ag._total(ag.current_phase_idx)
        txt = f"🟢 شارع {ag.current_phase_idx} مفتوح — {total} سيارة متبقية"
        col = (0,220,0)
    elif ag.sig == QLAgent.SIG_CLEARING:
        txt = f"⏳ تأكيد تفريغ شارع {ag.current_phase_idx}..."
        col = (0,200,200)
    else:
        txt = f"🟡 أصفر — {max(0,ag.phase_timer):.0f}ث"
        col = (0,200,200)

    cv2.putText(frame, txt, (bx+8,155), cv2.FONT_HERSHEY_SIMPLEX, 0.5, col, 2)

    cv2.putText(frame, "الاستراتيجية:",
                (bx+8,182), cv2.FONT_HERSHEY_SIMPLEX, 0.38, (180,180,100), 1)
    cv2.putText(frame, "احمر 5ث تجميع  ←  افتح الازحم  ←  استنى التفريغ",
                (bx+8,200), cv2.FONT_HERSHEY_SIMPLEX, 0.34, (140,140,140), 1)

    # زحمة الشوارع
    h_y = 228
    cv2.putText(frame, "زحمة الشوارع:",
                (bx+8, h_y-14), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (200,200,100), 1)
    for idx, _ in ag.green_phases:
        total   = ag._total(idx)
        waiting = ag._waiting(idx)
        is_open = (idx == ag.current_phase_idx and ag.sig in (QLAgent.SIG_GREEN, QLAgent.SIG_CLEARING))
        lbl     = f"شارع {idx}: كلي={total} | منتظر={waiting}" + (" ◀ مفتوح" if is_open else "")
        lcol    = (100,255,100) if is_open else (200,200,200)
        cv2.putText(frame, lbl, (bx+8, h_y), cv2.FONT_HERSHEY_SIMPLEX, 0.38, lcol, 1)

        bar = min(waiting*14, 300)
        cv2.rectangle(frame, (bx+8, h_y+6), (bx+8+300, h_y+14), (40,40,60), -1)
        cv2.rectangle(frame, (bx+8, h_y+6), (bx+8+bar,  h_y+14),
                      (50,220,50) if is_open else (50,120,220), -1)
        h_y += 40

    cv2.putText(frame, f"مرات التفريغ الكامل: {ag.cleared}",
                (bx+8, h_y+10), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (50,255,255), 1)

    # ── لوحة اليمين (السجل) ──
    rx = VIDEO_W-300
    cv2.rectangle(frame, (rx,100), (rx+288,430), (28,28,42), -1)
    cv2.rectangle(frame, (rx,100), (rx+288,430), (80,80,120), 1)
    cv2.putText(frame, "سجل القرارات",
                (rx+8,122), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,220,60), 1)
    for li, entry in enumerate(ag.log[-14:]):
        cv2.putText(frame, entry[:44], (rx+8, 148+li*19),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.31, (200,200,255), 1)

    return frame


def create_cfg():
    cfg = f"""<?xml version="1.0" encoding="UTF-8"?>
<configuration>
    <input>
        <net-file    value="project.net.xml"/>
        <route-files value="project.rou.xml"/>
    </input>
    <time>
        <begin value="0"/>
        <end   value="{SIM_STEPS}"/>
        <step-length value="{STEP_LEN}"/>
    </time>
</configuration>"""
    if not os.path.exists(SUMO_CFG):
        with open(SUMO_CFG, "w", encoding="utf-8") as f:
            f.write(cfg)


# ─────────────────────────────────────────────────────
#  التشغيل
# ─────────────────────────────────────────────────────
def run():
    create_cfg()
    sumo_bin = "sumo-gui"
    for c in [os.path.join(SUMO_HOME,"bin","sumo-gui.exe"),
              os.path.join(SUMO_HOME,"bin","sumo-gui")]:
        if os.path.exists(c):
            sumo_bin = c; break

    cmd = [sumo_bin, "-c", SUMO_CFG,
           "--step-length", str(STEP_LEN),
           "--no-warnings", "true",
           "--time-to-teleport", "-1",
           "--start", "--quit-on-end"]

    try:    traci.close()
    except: pass

    traci.start(cmd)
    tls_ids = traci.trafficlight.getIDList()
    if not tls_ids:
        traci.close(); return

    agents = {tid: QLAgent(tid) for tid in tls_ids}

    if os.path.exists(QTABLE_FILE):
        try:
            with open(QTABLE_FILE, "rb") as f:
                saved = pickle.load(f)
            for tid in agents:
                if tid in saved:
                    agents[tid].q_table = saved[tid]
            print("[✔] تم تحميل ملف التدريب.")
        except: pass

    fourcc = cv2.VideoWriter_fourcc(*"XVID")
    writer = cv2.VideoWriter(VIDEO_FILE, fourcc, VIDEO_FPS, (VIDEO_W, VIDEO_H))

    print("\n  [🚀] تشغيل v9.0 — أحمر 5ث تجميع ← الأكثر زحمة ← تفريغ كامل...")
    for step in range(SIM_STEPS):
        traci.simulationStep()
        for ag in agents.values():
            ag.step(step)
        if step % max(1, int(1/(VIDEO_FPS*STEP_LEN))) == 0:
            writer.write(draw_frame(step, agents))

    writer.release()
    traci.close()

    with open(QTABLE_FILE, "wb") as f:
        pickle.dump({tid: ag.q_table for tid, ag in agents.items()}, f)
    print("\n  [✔] انتهت المحاكاة — تم حفظ النتائج.")


if __name__ == "__main__":
    run()
```
